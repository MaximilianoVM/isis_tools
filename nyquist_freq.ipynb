```python
import numpy as np
```

```python
lc_path = '202407_I/593_I.dat'
```

```python
# Lee directamente en un array
data = np.loadtxt(lc_path)

# Cada columna en un array separado
date   = data[:,0].astype(float)
value   = data[:,1].astype(float)
```

```python
# Calcular las diferencias de tiempo entre observaciones consecutivas (en días)
time_differences = np.diff(date)

time_differences
```

```python
# Calcular el intervalo promedio de muestreo
average_interval = np.median(time_differences)

average_interval
```

```python
# Calcular la frecuencia de Nyquist (en ciclos/día)
nyquist_frequency = 0.5 * (1 / average_interval)

# También calcular el período de Nyquist (mínimo período detectable)
nyquist_period_days = 2 * average_interval

# Convertir a minutos para mejor interpretación
average_interval_minutes = average_interval * 1440  # 1 día = 1440 minutos
nyquist_period_minutes = nyquist_period_days * 1440

print("=== ANÁLISIS DE FRECUENCIA DE NYQUIST ===")
print(f"Número total de puntos: {len(date)}")
print(f"Intervalo promedio de muestreo: {average_interval:.6f} días")
print(f"Intervalo promedio: {average_interval_minutes:.2f} minutos")
print(f"Frecuencia de Nyquist: {nyquist_frequency:.2f} ciclos/día (c/d)")
print(f"Período de Nyquist (mínimo detectable): {nyquist_period_days:.6f} días")
print(f"Período de Nyquist: {nyquist_period_minutes:.2f} minutos")
print(f"Frecuencia máxima analizable: {nyquist_frequency:.2f} c/d")

# Información adicional útil
print(f"\n=== INFORMACIÓN ADICIONAL ===")
print(f"Rango temporal total: {date[-1] - date[0]:.3f} días")
print(f"Rango temporal en horas: {(date[-1] - date[0]) * 24:.1f} horas")
```
