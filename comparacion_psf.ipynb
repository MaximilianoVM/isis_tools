```python
header_file = 'NGC_6426I_2/header_info_images6426I_2'
data_file = 'NGC_6426I_2/lc210.data'
date_format = 'HJD'

header_info, lc_data = format_archives(headerfile_path=header_file, lc_data_path=data_file, date_format=date_format)
#display(lc_data)
```

```python
lc_data['HJD'] = lc_data['HJD'] + 2460400.0
```

```python
lc_data.head()
```

```python
# Configuración (ajusta estos valores)
h = 2460000.0       # Época de referencia (HJD)
p = 0.681990000000        # Período (ejemplo: 0.408133 días)
object_name = "V4323Oph"  # Nombre del objeto
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración (ajusta estos valores)
#h = 2460000.0       # Época de referencia (HJD)
#p = 0.408133        # Período (ejemplo: 0.408133 días)
#object_name = "V4324Oph"  # Nombre del objeto

# Función para calcular la fase y replicar los datos
def phase_replicate(hjd, flux, h, p):
    phase = ((hjd - h) / p) % 1.0  # Fase base [0, 1)
    # Replicar en [-1, 0), [0, 1), [1, 2)
    phases = np.concatenate([phase - 1.0, phase, phase + 1.0])
    fluxes = np.concatenate([flux, flux, flux])
    return phases, fluxes

# Calcular fases replicadas para I band
i_phases, i_fluxes = phase_replicate(lc_data['HJD'], lc_data['Flux'], h, p)

# Crear el gráfico
plt.figure(figsize=(10, 6))
plt.gca().invert_yaxis()  # Magnitudes más brillantes arriba

# Graficar datos I (rojo) con las tres repeticiones de fase
plt.scatter(i_phases, i_fluxes, s=10, color='red', label='I band', alpha=0.7)

# Ajustes estéticos (igual que en Gnuplot)
plt.xlim(-0.6, 0.6)
plt.xlabel('Phase')
plt.ylabel('Flux')  # Cambiado de Magnitude a Flux para coincidir con tus datos
plt.title(f'{object_name}. P={p}d, HJDo={h}')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```
